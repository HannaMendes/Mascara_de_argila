```python
# Contagem das Palavras

from bs4 import BeautifulSoup as bs
import requests

# Importação de conteudo do HTML da url.
res = requests.get("https://www.vibenaturais.com.br/beleza-e-bem-estar/cosmeticos-naturais/kit-argila-skin-care-com-9-cores-100g-cada")
soup = bs(res.text, 'html.parser')
texto = soup.get_text()

# Criação de variavies para seleção de palavras
substring =  "Argila"
substring1 = "Cosméticos"
substring2 = "Skin Care"

# Contator de palavras no contexto do HTML
count = texto.count(substring)
count1 = texto.count(substring1)
count2 = texto.count(substring2)

# print resultado da frequencia das palavras.
print(substring+" : ",  count)
print(substring1+" : ", count1)
print(substring2+" : ", count2)
```
